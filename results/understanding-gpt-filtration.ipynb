```python
import pandas as pd

# Function to filter dataframe based on a single percentile value across multiple columns
def filter_dataframe_by_single_percentile(df, columns, percentile):
    """
    Filters the dataframe by removing rows where any column value is below the specified percentile value.
    
    Parameters:
        df (pd.DataFrame): The dataframe to filter.
        columns (list of str): The list of column names to apply the filter on.
        percentile (float): The percentile threshold to apply (e.g., 0.2 for 20th percentile).
    
    Returns:
        pd.DataFrame: The filtered dataframe.
    """
    for column in columns:
        # Calculate the percentile value for the current column
        threshold = df[column].quantile(percentile)
        
        # Filter the dataframe to keep only rows where the column value is above or equal to the threshold
        df = df[df[column] >= threshold]
    
    return df

# Define the columns you want to apply the percentile filter to
columns_to_filter = ['class_representation', 'diversity', 'visual_fidelity', 'structural_integrity', 'beneficial_variations']  # Replace with your columns

# Define the percentile threshold (e.g., 0.2 for 20th percentile)
percentile_value = 0.1

# Apply the filter
filtered_df = filter_dataframe_by_single_percentile(df5, columns_to_filter, percentile_value)

filtered_df
```

```python
import pandas as pd
from scipy import stats

def filter_dataframe_by_zscore(df, columns, zscore_threshold=2):
    """
    Normalizes the data within each specified column using Z-score and filters out rows
    that are outside the specified Z-score threshold (rows where the absolute Z-score is greater than the threshold).
    
    Parameters:
        df (pd.DataFrame): The dataframe to filter.
        columns (list of str): The list of column names to apply Z-score normalization.
        zscore_threshold (float): The Z-score threshold to filter rows (e.g., 2 for 2 standard deviations from the mean).
    
    Returns:
        pd.DataFrame: The filtered dataframe.
    """
    # Create a copy of the dataframe to avoid modifying the original one
    df_filtered = df.copy()
    
    # Apply Z-score normalization to each specified column
    for column in columns:
        # Calculate Z-scores for the column
        df_filtered[column + '_zscore'] = stats.zscore(df_filtered[column])
        
        # Filter out rows where the absolute Z-score is greater than the threshold
        df_filtered = df_filtered[abs(df_filtered[column + '_zscore']) <= zscore_threshold]
    
    # Drop the added Z-score columns (optional, if you don't need them in the final output)
    df_filtered.drop([col + '_zscore' for col in columns], axis=1, inplace=True)
    
    return df_filtered

# Example usage:

# # Load your dataset
# file_path = 'your_dataset.csv'  # Replace with the actual file path to your dataset
# df = pd.read_csv(file_path)

# Define the columns you want to apply Z-score normalization and filtration to
columns_to_filter = ['class_representation', 'visual_fidelity', 'structural_integrity', 'diversity', 'beneficial_variations']  # Replace with your column names

# Define the Z-score threshold (e.g., 2 standard deviations from the mean)
zscore_threshold = 2

# Apply the Z-score normalization and filtration
filtered_df = filter_dataframe_by_zscore(df5, columns_to_filter, zscore_threshold)

len(filtered_df)
```

```python
import pandas as pd

def filter_dataframe_by_custom_score(df, threshold):
    """
    Creates a custom score by adding the values from 'class_representation', 'visual_fidelity', and 'structural_integrity',
    and multiplying by the sum of 'diversity' and 'beneficial_variations'. Filters the dataframe based on the threshold for this custom score.
    
    Parameters:
        df (pd.DataFrame): The dataframe to filter.
        threshold (float): The threshold value to filter rows based on the custom score.
    
    Returns:
        pd.DataFrame: The filtered dataframe with a new column 'custom_score'.
    """
    # Calculate the sum of 'class_representation', 'visual_fidelity', and 'structural_integrity'
    part1 = df['class_representation'] + df['visual_fidelity'] + df['structural_integrity']
    
    # Calculate the sum of 'diversity' and 'beneficial_variations'
    part2 = df['diversity'] + df['beneficial_variations']
    
    # Create the custom score by multiplying part1 by part2
    df['custom_score'] = part1 * part2
    
    # Filter the dataframe based on the threshold value for the custom score
    filtered_df = df[df['custom_score'] >= threshold]
    
    return filtered_df

# Example usage:

# # Load your dataset
# file_path = 'your_dataset.csv'  # Replace with the actual file path to your dataset
# df = pd.read_csv(file_path)

# Define the threshold for filtering based on the custom score
threshold_value = 1  # You can set any threshold value you want

# Apply the custom score filtration
filtered_df = filter_dataframe_by_custom_score(df5, threshold_value)

len(filtered_df)
```

```python
import pandas as pd
import numpy as np

# Load the data
df = df5

# Function to find edge cases
def find_edge_cases(df, criterion1, criterion2, threshold=0.3):
    diff = df[criterion1] - df[criterion2]
    return df[(diff > threshold) | (diff < -threshold)].sort_values(by=diff.abs().index, ascending=False)

# Edge cases: high class representation, low visual fidelity
print("High class representation, low visual fidelity:")
print(find_edge_cases(df, 'class_representation', 'visual_fidelity'))

# Edge cases: high visual fidelity, low class representation
print("\nHigh visual fidelity, low class representation:")
print(find_edge_cases(df, 'visual_fidelity', 'class_representation'))

# Edge cases: high visual fidelity, low structural integrity
print("\nHigh visual fidelity, low structural integrity:")
print(find_edge_cases(df, 'visual_fidelity', 'structural_integrity'))

# Edge cases: high structural integrity, low visual fidelity
print("\nHigh structural integrity, low visual fidelity:")
print(find_edge_cases(df, 'structural_integrity', 'visual_fidelity'))

# Edge cases: high diversity, low beneficial variations
print("\nHigh diversity, low beneficial variations:")
print(find_edge_cases(df, 'diversity', 'beneficial_variations'))

# Edge cases: high beneficial variations, low diversity
print("\nHigh beneficial variations, low diversity:")
print(find_edge_cases(df, 'beneficial_variations', 'diversity'))

# Calculate correlations
correlation_matrix = df[['class_representation', 'visual_fidelity', 'structural_integrity', 'diversity', 'beneficial_variations']].corr()
print("\nCorrelation Matrix:")
print(correlation_matrix)

# Function to find outliers
def find_outliers(df, column, threshold=1.5):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - (threshold * IQR)
    upper_bound = Q3 + (threshold * IQR)
    return df[(df[column] < lower_bound) | (df[column] > upper_bound)]

# Find outliers for each criterion
for column in ['class_representation', 'visual_fidelity', 'structural_integrity', 'diversity', 'beneficial_variations']:
    print(f"\nOutliers for {column}:")
    print(find_outliers(df, column))

# Summary statistics
print("\nSummary Statistics:")
print(df.describe())

# Distribution of scores
print("\nDistribution of scores (binned):")
for column in ['class_representation', 'visual_fidelity', 'structural_integrity', 'diversity', 'beneficial_variations']:
    print(f"\n{column}:")
    print(pd.cut(df[column], bins=10).value_counts().sort_index())

# Function to find specific edge cases
def find_specific_edge_cases(df, high_criterion, low_criterion, threshold=0.3):
    return df[(df[high_criterion] > df[low_criterion] + threshold) & 
              (df[low_criterion] < df[low_criterion].mean())].sort_values(by=high_criterion, ascending=False)

# Find specific edge cases
print("\nHigh class representation, low visual fidelity:")
print(find_specific_edge_cases(df, 'class_representation', 'visual_fidelity'))

print("\nHigh visual fidelity, low structural integrity:")
print(find_specific_edge_cases(df, 'visual_fidelity', 'structural_integrity'))

# Function to get top N diverse images for each class
def top_diverse_images(df, n=5):
    return df.sort_values('diversity', ascending=False).groupby('class_name').head(n)

print("\nTop 5 most diverse images for each class:")
print(top_diverse_images(df))

# Function to get images with high beneficial variations but low diversity
def high_variation_low_diversity(df, threshold=0.3):
    return df[(df['beneficial_variations'] > df['diversity'] + threshold) & 
              (df['diversity'] < df['diversity'].mean())].sort_values(by='beneficial_variations', ascending=False)

print("\nImages with high beneficial variations but low diversity:")
print(high_variation_low_diversity(df))
```
